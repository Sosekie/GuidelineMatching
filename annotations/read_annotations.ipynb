```python
from cassis import *
from pathlib import Path
from pprint import pprint
import pandas as pd
```

```python
document_annotation_path = Path("./annotations/1_Harmonisiert_oeffentliche_submission_gesunde_ernaehrung")
docname = "1_Harmonisiert_oeffentliche_submission_gesunde_ernaehrung.xmi"
```

```python
with open(document_annotation_path / "TypeSystem.xml", "rb") as f:
  typesystem = load_typesystem(f)

with open(document_annotation_path / docname, "rb") as f:
  doc = load_cas_from_xmi(f, typesystem=typesystem)
```

```python
named_entities = doc.select("de.tudarmstadt.ukp.dkpro.core.api.ner.type.NamedEntity")

pprint(named_entities[:3])
```

```python
data = {
    "text": [],
    "criterion_id": [],
    "criterion_type": []
}

for ne in named_entities:
    data["criterion_id"].append(ne.identifier.split("#")[-1])
    data["criterion_type"].append(ne.value)
    data["text"].append(ne.get_covered_text())

df = pd.DataFrame(data)

df
```
